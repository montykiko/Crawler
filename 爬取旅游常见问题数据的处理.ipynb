# 帮助中心

```python
data = pd.read_table('content.txt',encoding = 'utf-8',header = None)
data
```

```python
data.columns = ['question']
data
```

```python
cate_list = ['会员服务', '外币兑换', '国内酒店', '海外酒店', '国内机票', '国际机票', '旅游度假', '机+酒', '火车票', '门票', 'Android Wear引导教程']
```

```python
for cate in cate_list:
    data.loc[data['question']==cate,'category'] = cate
```

```python
data
```

```python
data.fillna(method = 'ffill',inplace = True)
```

```python
df1 = data[['category','question']].copy()
df1
```

```python
df1.shape
```

```python
df1.drop_duplicates(inplace = True)
```

```python
df1.shape
```

```python
df1
```

```python
cond = df1['category'] == df1['question']
```

```python
df2 = df1[~cond]
```

```python
df2.index = range(len(df2))
```

```python
df2
```

```python
df2.to_csv('快速索引.txt',sep = '\t',encoding = 'utf-8')
```
